```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Extracting Data from txt file

```python
data = np.genfromtxt("HR data.txt" , delimiter = ',' ,dtype = object)
new_data = []

for row in data[:]:
    x = row.split()
    if len(x) < 9:
        continue
    else:
        new_data.append([value for value in x])
```

## Extracting VMag , PLX , B-V

## Generating Absolute Magnitude using the formula:
###  $$Absolute Magnitude = Apparent Magnitude + 5\log\left(\frac{PLX}{1000}\right) + 5$$

```python
vmag = []
plx = []
b_v = []
mv = []

for i in range(1,len(new_data)):
    vmag.append(float(new_data[i][1]))
    plx.append(float(new_data[i][4]))
    b_v.append(float(new_data[i][8]))
    mv.append(vmag[i-1] +( 5 * np.log10(plx[i-1]/1000)) + 5)
```

## Plotting HR Diagram

```python
plt.figure(figsize = [16,10])
plt.scatter(y = mv , x = b_v , color = 'white')
plt.style.use('dark_background')
plt.ylim(10 , -4)
plt.xlabel('B-V (Color Index)' , size = 20)
plt.ylabel('Absolute Magnitude' , size = 20)
plt.title('Hertzsprung-Russell Diagram' , size = 35)
plt.show()
```

### This 2nd plot takes awhile to run due to brute force used for colouring

```python

for i in range(len(b_v)):
    if  0 > b_v[i] :
         plt.scatter(b_v[i], mv[i], color='#003dc2')
    elif  0 <= b_v[i] < 0.5 :
         plt.scatter(b_v[i], mv[i], color='#a6a6c9')
    elif  0.5 <= b_v[i] < 1 :
         plt.scatter(b_v[i], mv[i], color='#ffff80')
    elif  1 <= b_v[i] < 1.5 :
         plt.scatter(b_v[i], mv[i], color='#ff5c00')
    elif  1.5 <= b_v[i] < 2 :
         plt.scatter(b_v[i], mv[i], color='#f22600')
    elif  2 <= b_v[i] :
         plt.scatter(b_v[i], mv[i], color='#cc0000')
            
plt.style.use('dark_background')
plt.ylim(10 , -4)
plt.xlabel('B-V (Color Index)' , size = 10)
plt.ylabel('Absolute Magnitude' , size = 10)
plt.title('Hertzsprung-Russell Diagram' , size = 20)
plt.show()
```
